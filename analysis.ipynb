```python
import pandas as pd
import numpy as np
print("Libraries loaded!")
```

```python
df = pd.read_csv('Combined_dataset.csv')
print("Data loaded successfully!")
print("Shape:", df.shape)
```

```python
# Pehle 5 rows
print("=== FIRST 5 ROWS ===")
print(df.head())

# Last 5 rows
print("\n=== LAST 5 ROWS ===")
print(df.tail())

# Shape
print("\n=== SHAPE ===")
print(df.shape)

# Columns
print("\n=== COLUMNS ===")
print(df.columns.tolist())

# Data Types
print("\n=== DATA TYPES ===")
print(df.dtypes)
```

```python
# Info - column types aur missing values
print("=== INFO ===")
df.info()

# Describe - statistical summary
print("\n=== DESCRIBE ===")
print(df.describe())
```

```python
print("=== MISSING VALUES ===")
print(df.isnull().sum())

print("\n=== MISSING VALUES % ===")
print((df.isnull().sum() / len(df) * 100).round(2))
```

```python
# Text columns mein "Unknown" bharo
df['what_customers_said'] = df['what_customers_said'].fillna('Unknown')
df['seller_name'] = df['seller_name'].fillna('Unknown')
df['seller_information'] = df['seller_information'].fillna('Unknown')
df['videos'] = df['videos'].fillna('No Video')
df['variations'] = df['variations'].fillna('No Variations')

# Check karo ab missing values hain ya nahi
print("=== MISSING VALUES AFTER CLEANING ===")
print(df.isnull().sum())
print("\n✅ Missing values handled!")
```

```python
# Discount ki missing values median se bharo
df['discount'] = df['discount'].fillna(df['discount'].median())

print("✅ Discount column fixed!")

# Filter — sirf wo products jo 4+ rating wale hain
high_rated = df[df['rating'] >= 4.0]
print(f"\n=== HIGH RATED PRODUCTS (4.0+) ===")
print(f"Total products: {len(high_rated)}")
print(high_rated[['title', 'rating', 'initial_price', 'category']].head(10))
```

```python
print("Before duplicates removal:", df.shape)

df = df.drop_duplicates()
df = df.reset_index(drop=True)

print("After duplicates removal:", df.shape)
print("✅ Duplicates removed!")
```

```python
# total_amount = initial_price * ratings_count
df['total_amount'] = df['initial_price'] * df['ratings_count']

print("✅ total_amount column created!")
print("\n=== SAMPLE DATA ===")
print(df[['title', 'initial_price', 'ratings_count', 'total_amount']].head(10))
```

```python
df.to_csv('cleaned_dataset.csv', index=False)

print("✅ File saved successfully!")
print("Final shape:", df.shape)
print("File saved as: cleaned_dataset.csv")
```
